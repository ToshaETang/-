```python
m='新美國小-13-2-d1'

df9=pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\816\濾過\{}_clean.csv'.format(m))

df9['datetime']=pd.to_datetime(df9['datetime'])        
df9.set_index('datetime', drop=True, inplace=True)        
df9=df9.resample(rule='30T').mean().interpolate(method='slinear')

try:
    res= STL(df9.distance1,period=103).fit()
except:
    res= STL(df9.distance2,period=103).fit()
    
# print(res.seasonal)
res.plot()
plt.show()

df9['trend']=res.trend
df9['season']=res.seasonal
df9['residual']=res.resid
#補滿前後資料
df9=df9.fillna(method='ffill')
df9=df9.fillna(method='bfill')
#print(name+"資料STL處理完成")
df9.to_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\816\{}_STL.csv'.format(m))
print(m)
```
