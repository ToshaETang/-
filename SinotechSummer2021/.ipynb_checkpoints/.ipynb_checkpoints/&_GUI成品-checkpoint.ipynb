```python
import matplotlib.pyplot as plt
from tkinter import *
import tkinter.ttk as ttk
import tkinter as tk
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk
import pandas as pd
import numpy as np
import os
import os.path
from os import listdir
from os.path import isfile, isdir, join
import datetime,time
import plotly.express as px
from tkinter import filedialog
import statsmodels
from pandas.plotting import register_matplotlib_converters  
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.seasonal import STL
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import openpyxl
import pytz
from datetime import datetime
from datetime import timedelta
import stl
from dateutil.parser import parse
from dateutil.tz import tzutc
from tzlocal import get_localzone
import plotly.offline as offline
import matplotlib
from tkinter import filedialog
from matplotlib.figure import Figure
from tkinter.filedialog import askdirectory
from PIL import Image, ImageTk
#----去掉紅色警告-------
from sklearn import metrics
import warnings
warnings.filterwarnings("ignore")
#==========================================================================================================
root0 = Tk()
root0.minsize(250, 300)
root0.title('主頁面')
#-----------------------------------------------------------------------------雨場切割---------------------
#做雨場切割分頁
def createroot2():                     
    root2 = tk.Toplevel()
    root2.minsize(1300, 700)
    root2.title('雨場切割')

    first = []
    end = []
    first_new = []
    test = []
    time = []
    
    #【選擇自己測的原始降雨資料】按鈕           
    def hit_raincut1():                         
        root0.withdraw()
        
        Labe00=tk.Label(root2,text='選擇資料->輸入資料中時間、時雨量column名稱')
        Labe00.place(x=510, y=70, width=500, height=30)
        
        file = filedialog.askopenfilename(title='選擇自己測的原始降雨資料', filetypes=[('Excel', '*.xlsx')])   #選檔案
        varr = tk.StringVar()
        entryr = tk.Entry(root2, width=20, textvariable=varr)
        entryr.insert(0, "時雨量column")
        entryr.place(x=630, y=37)
        
        vart = tk.StringVar()
        entryt = tk.Entry(root2, width=20, textvariable=vart)
        entryt.insert(0, "時間column")
        entryt.place(x=630, y=7)
        
        #【取得columns】按鈕
        def get():                                                    
            df = pd.read_excel(r'{}'.format(file))                              
            varr = entryr.get()
            test = df['{}'.format(varr)].values.ravel()        #抓輸入的column名
            vart = entryt.get()
            df.set_index("{}".format(vart), inplace=True)
            df.index = pd.DatetimeIndex(df.index)
            df = df.resample('H').sum()
            time = df.index.values.ravel() 
            time = pd.to_datetime(pd.Series(time))
        
            for i in range(len(time)):
                time[i] = time[i].replace(tzinfo=None)          #去掉時區
    
            for i in range(len(test)):
                if test[i] > 4 :
                    first.append(int(i))
                    x = test[i + 1 : i + 7]
                    if all(j <= 4 for j in x): 
                        end.append(int(i) + 1)
                    else:
                        continue
            if len(first)>0:      
    #判斷降雨事件間隔是否有大於6小時
                first_new.append(first[0])
                for i in range(len(first)):
                    if (first[i] - first[i - 1]) > 6:
                        first_new.append(first[i])
                        rain_result = []
                        rain_sum = []
                        rain_max = []
                        time_first = []
                        time_end = []
                        time_lag = []
    #各個欄位計算
            for i in range(len(first_new)):
                timelag = end[i] - first_new[i]
                time_first.append(time[first_new[i]] + timedelta(hours=8))   #加8小時回去
                time_end.append(time[end[i]] + timedelta(hours=8))
                rain_result.append(test[first_new[i] : end[i] + 1])
                rain_sum.append(sum(test[first_new[i] : end[i] + 1]))
                time_lag.append(timelag)
            for i in range(len(rain_result)):
                rain_max.append(max(rain_result[i]))
            for i in range(len(rain_result)):
                rain_result[i]=str(rain_result[i])
                df_L_rain = pd.DataFrame(np.array([rain_result]))
        
    #有判斷欄位的原始雨場
            result = pd.DataFrame()
            index = 0
            for i in range(len(rain_result)):
                if rain_sum[i] > 12.7:
                    d = {"Index":index,
                         "ID":"try",
                         "S_time":time_first[i],
                         "E_time":time_end[i],
                         "降雨延時":time_lag[i]+1,
                         "總降雨量":rain_sum[i],
                         "最大降雨量":rain_max[i],
                         "rain_detail":df_L_rain[i],
                         "sum是否大於12.7":"1"
                         }
                    df_test = pd.DataFrame(data = d, index=[0])
                    result = pd.concat([result, df_test],axis =0)
                    index = index + 1
                else:
                    d = {"Index":index,
                         "ID":"try",
                         "S_time":time_first[i],
                         "E_time":time_end[i],
                         "降雨延時":time_lag[i]+1,
                         "總降雨量":rain_sum[i],
                         "最大降雨量":rain_max[i],
                         "rain_detail":df_L_rain[i],
                         "sum是否大於12.7":"0"
                         }
                    df_test = pd.DataFrame(data = d,index=[0])
                    result = pd.concat([result, df_test],axis =0)
                    index = index + 1
                
                    #存檔
                    var = tk.StringVar()
                    entry = tk.Entry(root2, width=20, textvariable=var)
                    entry.insert(0, "XXX .xlsx")     #檔名提示                  
                    entry.place(x=230, y=7)
                    Labe00=tk.Label(root2,text='輸入檔名(要有檔案類型)->選擇存檔資料夾')
                    Labe00.place(x=95, y=40, width=500, height=30)
                    def save():
                        var = entry.get()
                        spath = filedialog.askdirectory(title='選擇存檔資料夾')
                        result.to_excel(r'{}'.format(os.path.join(spath,str(var))))
                        Label=tk.Label(root2,text='雨場切割已儲存')
                        Label.place(x=100, y=5, width=100, height=30)
                        LabeL2=tk.Label(root2,text='共有{}筆資料'.format(len(result)))
                        LabeL2.place(x=20, y=155, width=200, height=30)
                        LabeL3=tk.Label(root2,text='{}'.format(str(result['S_time'])))
                        LabeL3.place(x=10, y=190, width=300, height=500)    
                    bsave = tk.Button(root2,text='輸入檔名',command=save)
                    bsave.place(x=330, y=5)
                
        bget = tk.Button(root2,text='取得columns',command=get)
        bget.place(x=850, y=10)                    
    
    #【選擇氣象站原始降雨資料】按鈕    氣象站資料clumn的名字固定，沒有時區問題
    def hit_raincut2():
        root0.withdraw()   #讓主畫面下移
        file = filedialog.askopenfilename(title='選擇氣象站原始降雨資料', filetypes=[('csv', '*.csv')])
        df = pd.read_csv(r'{}'.format(file))
        df.set_index("時間", inplace=True)
        df.index = pd.DatetimeIndex(df.index)
        df = df.resample('H').sum()
        
        test = df['時雨量'].values.ravel() 
        time = df.index.values.ravel() 
        time = pd.to_datetime(pd.Series(time))
    
        for i in range(len(time)):
            time[i] = time[i].replace(tzinfo=None)   #(雖然沒差但還是寫一下)
    
        for i in range(len(test)):
            if test[i] > 4 :
                first.append(int(i))
                x = test[i + 1 : i + 7]
                if all(j <= 4 for j in x): 
                    end.append(int(i) + 1)
                else:
                    continue
        if len(first)>0:
    #判斷降雨事件間隔是否有大於6小時
            first_new.append(first[0])
            for i in range(len(first)):
                if (first[i] - first[i - 1]) > 6:
                    first_new.append(first[i])
                    rain_result = []
                    rain_sum = []
                    rain_max = []
                    time_first = []
                    time_end = []
                    time_lag = []
    #各個欄位計算
        for i in range(len(first_new)):
            timelag = end[i] - first_new[i]
            time_first.append(time[first_new[i]]+ timedelta(hours=8))
            time_end.append(time[end[i]]+ timedelta(hours=8))
            rain_result.append(test[first_new[i] : end[i] + 1])
            rain_sum.append(sum(test[first_new[i] : end[i] + 1]))
            time_lag.append(timelag)
        for i in range(len(rain_result)):
            rain_max.append(max(rain_result[i]))
        for i in range(len(rain_result)):
            rain_result[i]=str(rain_result[i])
            df_L_rain = pd.DataFrame(np.array([rain_result]))
        
    #有判斷欄位的原始雨場
        result = pd.DataFrame()
        index = 0
        for i in range(len(rain_result)):
            if rain_sum[i] > 12.7:
                d = {"Index":index,
                     "ID":"try",
                     "S_time":time_first[i],
                     "E_time":time_end[i],
                     "降雨延時":time_lag[i]+1,
                     "總降雨量":rain_sum[i],
                     "最大降雨量":rain_max[i],
                     "rain_detail":df_L_rain[i],
                     "sum是否大於12.7":"1"
                     }
                df_test = pd.DataFrame(data = d,index=[0])
                result = pd.concat([result, df_test],axis =0)
                index = index + 1
            else:
                d = {"Index":index,
                     "ID":"try",
                     "S_time":time_first[i],
                     "E_time":time_end[i],
                     "降雨延時":time_lag[i]+1,
                     "總降雨量":rain_sum[i],
                     "最大降雨量":rain_max[i],
                     "rain_detail":df_L_rain[i],
                     "sum是否大於12.7":"0"
                     }
                df_test = pd.DataFrame(data = d,index=[0])
                result = pd.concat([result, df_test],axis =0)
                index = index + 1
        
        #存檔
        var = tk.StringVar()
        entry = tk.Entry(root2, width=20, textvariable=var)
        entry.insert(0, "XXX .xlsx")    #檔名範例
        entry.place(x=230, y=67)
        Labe00=tk.Label(root2,text='輸入檔名(要有檔案類型)->選擇存檔資料夾')
        Labe00.place(x=95, y=100, width=500, height=30)
        def save():
            var = entry.get()
            spath = filedialog.askdirectory(title='選擇存檔資料夾')
            result.to_excel(r'{}'.format(os.path.join(spath,str(var))))
            Label=tk.Label(root2,text='雨場切割已儲存')
            Label.place(x=100, y=60, width=100, height=30)
            LabeL2=tk.Label(root2,text='共有{}筆資料'.format(len(result)))
            LabeL2.place(x=9, y=155, width=200, height=30)
            LabeL3=tk.Label(root2,text='{}'.format(str(result['S_time'])))
            LabeL3.place(x=10, y=190, width=300, height=500)
        bsave = tk.Button(root2,text='輸入檔名',command=save)
        bsave.place(x=330, y=65)
    
    #回主畫面
    def Bm():
        root2.destroy()
        root0.deiconify()

    raincut1 = tk.Button(root2, text='自己監測資料', bg='FireBrick',fg='white', state=tk.NORMAL, command=hit_raincut1)
    raincut1.place(x=10, y=5, width=80, height=30)
    raincut2 = tk.Button(root2, text='氣象站資料', bg='FireBrick',fg='white', state=tk.NORMAL, command=hit_raincut2)
    raincut2.place(x=10, y=60, width=80, height=30)      
    Bm = tk.Button(root2, text='返回主畫面', bg='WhiteSmoke',fg='gray', state=tk.NORMAL, command=Bm)   
    Bm.place(x=10, y=115, width=80, height=30)
#---------------------------------------------------------------------------------坡面侵蝕-------------------
#做坡面侵蝕分頁
def createroot():
    root = tk.Toplevel()    #做彈出視窗
    root.minsize(1300, 700)
    root.title('坡面侵蝕')
    spath = filedialog.askdirectory(title='選擇存檔資料夾')  #因為生成的檔案很多，且檔名固定不變，故直接在一開始選擇存檔位置
    
    #【切割ID】按鈕
    def hit_split1():
        global m1,m2
        root0.withdraw()
        file = filedialog.askopenfilename(title='選擇原始檔案', filetypes=[('csv', '*.csv')])
        df0 = pd.read_csv(r'{}'.format(file))
        df1 = df0.loc[df0['2']!=20]    #去掉距離為20的錯誤資料                    
        
        #做原始資料的圖
        res1= STL(df1['2'],period=103).fit()
        res1.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"merge_d1.png")))    #圖片存入一開始所選的資料夾
        res2= STL(df1['3'],period=103).fit()
        res2.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"merge_d2.png")))    #圖片存入一開始所選的資料夾
    
        m1=Image.open(r'{}'.format(os.path.join(spath,"merge_d1.png")))  #開啟圖片
        m1=m1.resize((252,224),Image.ANTIALIAS)
        m1=ImageTk.PhotoImage(m1)
        m1Label=tk.Label(root,image=m1)     #放圖片到介面                    
        m1Label.place(x=550, y=20, width=252, height=224)
        mn1Label=tk.Label(root,text='原始資料_d1')
        mn1Label.place(x=640, y=5, width=100, height=30)
    
        m2=Image.open(r'{}'.format(os.path.join(spath,"merge_d2.png")))
        m2=m2.resize((252,224),Image.ANTIALIAS)
        m2=ImageTk.PhotoImage(m2)
        m2Label=tk.Label(root,image=m2)
        m2Label.place(x=900, y=20, width=252, height=224)
        mn2Label=tk.Label(root,text='原始資料_d2')
        mn2Label.place(x=990, y=5, width=100, height=30)
        
        #用設備編號(1&2)切資料
        df_id1=df0.loc[df0['1']==1]
        df_id1.to_csv(r'{}'.format(os.path.join(spath,"ID1.csv")))
        df_id2=df0.loc[df0['1']==2]
        df_id1.to_csv(r'{}'.format(os.path.join(spath,"ID2.csv")))
        Label = tk.Label(root, text='原始資料已分割成ID1、ID2儲存')
        Label.place(x=100, y=0, width=180, height=30)
    
    #【切割距離】按鈕
    def hit_split2():
        df01 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID1.csv")))
        df02 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID2.csv")))
        df01 = df01.set_index('Time')
        df02 = df02.set_index('Time') 
        df11 = df01.drop(['Unnamed: 0','Device Id','1','3'],axis=1).rename(columns={'2':'distance1'})
        df12 = df01.drop(['Unnamed: 0','Device Id','1','2'],axis=1).rename(columns={'3':'distance2'})        
        df11.to_csv(r'{}'.format(os.path.join(spath,"ID1_d1.csv")))
        df12.to_csv(r'{}'.format(os.path.join(spath,"ID1_d2.csv")))
        df21 = df02.drop(['Unnamed: 0','Device Id','1','3'],axis=1).rename(columns={'2':'distance1'})
        df22 = df02.drop(['Unnamed: 0','Device Id','1','2'],axis=1).rename(columns={'3':'distance2'})        
        df21.to_csv(r'{}'.format(os.path.join(spath,"ID2_d1.csv")))
        df22.to_csv(r'{}'.format(os.path.join(spath,"ID2_d2.csv")))
        Label = tk.Label(root, text='ID1、ID2已分割ID1_d1、ID1_d2、ID2_d1、ID2_d2成儲存')
        Label.place(x=100, y=50, width=330, height=30)
    
    #【做STL】按鈕
    def hit_dostl():
        global img1,img2,img3,img4
        df01 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID1_d1.csv")))
        df02 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID1_d2.csv")))
        df03 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID2_d1.csv")))
        df04 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID2_d2.csv")))
        #-------------------------------------------------------------------------------清洗資料--------
        df02['Time']=pd.to_datetime(df02['Time'])
        for i in range(len(df02)):
            df02.Time[i] = df02.Time[i].replace(tzinfo=None)
            df02.Time[i] = df02.Time[i] + timedelta(hours=8) 
        df02 = df02.loc[df02['distance2']!=20]
        df02.set_index('Time', drop=True, inplace=True)
        df02=df02.loc[df02["distance2"]>=30]["distance2"]
        mean=df02.rolling(30,min_periods=1,center=True).mean()
        std=df02.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df02)):
            if abs(df02[i]-mean[i])<std[i]:
                i=i+1
            else:
                df02=df02.drop(df02.index[i])
                i=i+1
        df2=df02.resample(rule='30T').mean()  
        mean=df2.rolling(30,min_periods=1,center=True).mean()
        std=df2.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df2)):
            if abs(df2[i]-mean[i])<std[i]:
                i=i+1
            else:
                df2=df2.drop(df2.index[i])
                i=i+1
        df3=df2.resample(rule='30T').mean().interpolate(method='slinear')
        df3.dropna(inplace=True)
        df=df3.to_frame()
        df.to_csv(r'{}'.format(os.path.join(spath,"ID1_d2.csv")))
        
        df01['Time']=pd.to_datetime(df01['Time'])
        for i in range(len(df01)):
            df01.Time[i] = df01.Time[i].replace(tzinfo=None)
            df01.Time[i] = df01.Time[i] + timedelta(hours=8)
        df01 = df01.loc[df01['distance1']!=20]
        df01.set_index('Time', drop=True, inplace=True)
        df01=df01.loc[df01["distance1"]>=30]["distance1"]
        mean=df01.rolling(30,min_periods=1,center=True).mean()
        std=df01.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df01)):
            if abs(df01[i]-mean[i])<std[i]:
                i=i+1
            else:
                df01=df01.drop(df01.index[i])
                i=i+1
        df2=df01.resample(rule='30T').mean()
        mean=df2.rolling(30,min_periods=1,center=True).mean()
        std=df2.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df2)):
            if abs(df2[i]-mean[i])<std[i]:
                i=i+1
            else:
                df2=df2.drop(df2.index[i])
                i=i+1
        df3=df2.resample(rule='30T').mean().interpolate(method='slinear')
        df3.dropna(inplace=True)
        df=df3.to_frame()
        df.to_csv(r'{}'.format(os.path.join(spath,"ID1_d1.csv")))
        
        df03['Time']=pd.to_datetime(df03['Time'])
        for i in range(len(df03)):
            df03.Time[i] = df03.Time[i].replace(tzinfo=None)
            df03.Time[i] = df03.Time[i] + timedelta(hours=8)
        df03 = df03.loc[df03['distance1']!=20]
        df03.set_index('Time', drop=True, inplace=True)
        df03=df03.loc[df03["distance1"]>=30]["distance1"]
        mean=df03.rolling(30,min_periods=1,center=True).mean()
        std=df03.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df03)):
            if abs(df03[i]-mean[i])<std[i]:
                i=i+1
            else:
                df03=df03.drop(df03.index[i])
                i=i+1
        df2=df03.resample(rule='30T').mean()   
        mean=df2.rolling(30,min_periods=1,center=True).mean()
        std=df2.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df2)):
            if abs(df2[i]-mean[i])<std[i]:
                i=i+1
            else:
                df2=df2.drop(df2.index[i])
                i=i+1
        df3=df2.resample(rule='30T').mean().interpolate(method='slinear')
        df3.dropna(inplace=True)
        df=df3.to_frame()
        df.to_csv(r'{}'.format(os.path.join(spath,"ID2_d1.csv")))
        
        df04['Time']=pd.to_datetime(df04['Time'])
        for i in range(len(df04)):
            df04.Time[i] = df04.Time[i].replace(tzinfo=None)
            df04.Time[i] = df04.Time[i] + timedelta(hours=8)
        df04 = df04.loc[df04['distance2']!=20]
        df04.set_index('Time', drop=True, inplace=True)
        df04=df04.loc[df04["distance2"]>=30]["distance2"]
        mean=df04.rolling(30,min_periods=1,center=True).mean()
        std=df04.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df04)):
            if abs(df04[i]-mean[i])<std[i]:
                i=i+1
            else:
                df04=df04.drop(df04.index[i])
                i=i+1
        df2=df04.resample(rule='30T').mean()
        mean=df2.rolling(30,min_periods=1,center=True).mean()
        std=df2.rolling(30,min_periods=1,center=True).std()
        i=0
        while i < (len(df2)):
            if abs(df2[i]-mean[i])<std[i]:
                i=i+1
            else:
                df2=df2.drop(df2.index[i])
                i=i+1
        df3=df2.resample(rule='30T').mean().interpolate(method='slinear')
        df3.dropna(inplace=True)
        df=df3.to_frame()
        df.to_csv(r'{}'.format(os.path.join(spath,"ID2_d2.csv")))
    #----------------------------------------------------------------------------------------做STL---------
        df01 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID1_d1.csv")))
        df02 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID1_d2.csv")))
        df03 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID2_d1.csv")))
        df04 = pd.read_csv(r'{}'.format(os.path.join(spath,"ID2_d2.csv")))
        res1= STL(df01.distance1,period=103).fit()
        res1.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"ID1_d1_stl.png")))
        df01['trend']=res1.trend
        df01['season']=res1.seasonal
        df01['residual']=res1.resid
        df01=df01.fillna(method='ffill')
        df01=df01.fillna(method='bfill')
        df01.to_csv(r'{}'.format(os.path.join(spath,"ID1_d1_stl.csv")))
        res2= STL(df02.distance2,period=103).fit()
        res2.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"ID1_d2_stl.png")))
        df02['trend']=res2.trend
        df02['season']=res2.seasonal
        df02['residual']=res2.resid
        df02=df02.fillna(method='ffill')
        df02=df02.fillna(method='bfill')
        df02.to_csv(r'{}'.format(os.path.join(spath,"ID1_d2_stl.csv")))
        res3= STL(df03.distance1,period=103).fit()
        res3.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"ID2_d1_stl.png")))
        df03['trend']=res3.trend
        df03['season']=res3.seasonal
        df03['residual']=res3.resid
        df03=df03.fillna(method='ffill')
        df03=df03.fillna(method='bfill')
        df03.to_csv(r'{}'.format(os.path.join(spath,"ID2_d1_stl.csv")))
        res4= STL(df04.distance2,period=103).fit()
        res4.plot()
        plt.savefig(r'{}'.format(os.path.join(spath,"ID2_d2_stl.png")))
        df04['trend']=res4.trend
        df04['season']=res4.seasonal
        df04['residual']=res4.resid
        df04=df04.fillna(method='ffill')
        df04=df04.fillna(method='bfill')
        df04.to_csv(r'{}'.format(os.path.join(spath,"ID2_d2_stl.csv")))
    #--------------------------------------------------------------------------------------輸出圖--------
        img1=Image.open(r'{}'.format(os.path.join(spath,"ID1_d1_stl.png")))
        img1=img1.resize((315,280),Image.ANTIALIAS)
        img1=ImageTk.PhotoImage(img1)
        img1Label=tk.Label(root,image=img1)
        img1Label.place(x=10, y=260, width=315, height=280)
        imgn1Label=tk.Label(root,text='ID1_d1')
        imgn1Label.place(x=150, y=535, width=50, height=30)    
        img2=Image.open(r'{}'.format(os.path.join(spath,"ID1_d2_stl.png")))
        img2=img2.resize((315,280),Image.ANTIALIAS)
        img2=ImageTk.PhotoImage(img2)
        img2Label=tk.Label(root,image=img2)
        img2Label.place(x=330, y=260, width=315, height=280)
        imgn2Label=tk.Label(root,text='ID1_d2')
        imgn2Label.place(x=470, y=535, width=50, height=30)   
        img3=Image.open(r'{}'.format(os.path.join(spath,"ID2_d1_stl.png")))
        img3=img3.resize((315,280),Image.ANTIALIAS)
        img3=ImageTk.PhotoImage(img3)
        img3Label=tk.Label(root,image=img3)
        img3Label.place(x=650, y=260, width=315, height=280)
        imgn3Label=tk.Label(root,text='ID2_d1')
        imgn3Label.place(x=790, y=535, width=50, height=30)    
        img4=Image.open(r'{}'.format(os.path.join(spath,"ID2_d2_stl.png")))
        img4=img4.resize((315,280),Image.ANTIALIAS)
        img4=ImageTk.PhotoImage(img4)
        img4Label=tk.Label(root,image=img4)
        img4Label.place(x=970, y=260, width=315, height=280)
        imgn4Label=tk.Label(root,text='ID2_d2')
        imgn4Label.place(x=1110, y=535, width=50, height=30)
        
    def Bm():
        root.destroy()
        root0.deiconify()
  
    split1 = tk.Button(root, text='切割ID', bg='Navy',fg='white', state=tk.NORMAL, command=hit_split1)     #分割原始資料(ID)
    split1.place(x=10, y=5, width=80, height=30)
    split2 = tk.Button(root, text='切割距離', bg='Navy',fg='white', state=tk.NORMAL, command=hit_split2)     #分割距離(d)
    split2.place(x=10, y=50, width=80, height=30)
    dostl = tk.Button(root, text='做STL', bg='Navy',fg='white',state=tk.NORMAL, command=hit_dostl)           #做stl
    dostl.place(x=10, y=95, width=80, height=30)
    Bm = tk.Button(root, text='返回主畫面', bg='WhiteSmoke',fg='gray', state=tk.NORMAL, command=Bm)   
    Bm.place(x=10, y=135, width=80, height=30)    
#--------------------------------------------------------------------------------沖蝕分析-------    
#做沖蝕分析分頁
def createroot3():
    root3 = tk.Toplevel()   
    root3.minsize(1300, 700)
    root3.title('沖蝕分析')
    
    #【沖蝕分析】按鈕
    def Erosion():
        root0.withdraw()
        file1 = filedialog.askopenfilename(title='選擇雨場分析資料', filetypes=[('Excel', '*.xlsx'),('Excel', '*.xls'),('csv', '*.csv')])
        rain = pd.read_excel(r'{}'.format(file1))
        file2 = filedialog.askopenfilename(title='選擇坡面侵蝕資料', filetypes=[('csv', '*.csv'),('Excel', '*.xlsx')])
        ero = pd.read_csv(r'{}'.format(file2))

        for i in range(len(rain)):
            rain.S_time[i]=datetime.strptime(str(rain.S_time[i]), "%Y-%m-%d %H:%M:%S")
            rain.E_time[i]=datetime.strptime(str(rain.E_time[i]), "%Y-%m-%d %H:%M:%S")
               
        for i in range(len(ero)):
            ero.time[i] = datetime.strptime(ero.time[i],"%Y/%m/%d %H:%M")    
            ero.time[i] = ero.time[i].replace(tzinfo=None)
            rain = rain.drop(rain.loc[rain.E_time<=ero.time[0]].index).reset_index(drop=True)
    
        Erosion_s = []
        Erosion_e = []
        time_record=[]
        
        for k in range(len(ero.time)):
            for w in range(len(rain)):                        
                k=2
                while k<len(ero.time):
                    if ero.time[k] > rain.S_time[w]:                                            
                        Erosion_s.append(ero.trend[k-2].round(2))
                        while k<len(ero.time):
                            if  ero.time[k]> rain.E_time[w]:  
                                Erosion_e.append(ero.trend[k].round(2))
                                time_record.append(rain.E_time[w])
                                break
                            else:                       
                                k+=1
                        break
                    else:
                        k+=1
                continue                       
            break

        d= {'Erosion_start':Erosion_s,               
                'Erosion_end':Erosion_e,
                'E_time':time_record}
        cc=pd.DataFrame(d)
        cc['E_time']=cc['E_time'].astype('datetime64[ns]')
        rain['E_time']=rain['E_time'].astype('datetime64[ns]')
        rain_all = pd.merge(rain,cc,on='E_time').drop(['index','Unnamed: 0'],axis=1)
        rain_all['Erosion']=rain_all['Erosion_end']-rain_all['Erosion_start']

        for i in range(len(rain_all)):                                                #去掉時區
            rain_all['S_time'][i] = rain_all['S_time'][i].replace(tzinfo=None)
            rain_all['E_time'][i] = rain_all['E_time'][i].replace(tzinfo=None)
            rain_all['S_time'][i] = rain_all['S_time'][i] + timedelta(hours=8)
            rain_all['S_time'][i] = rain_all['S_time'][i] + timedelta(hours=8)
        
        #存檔
        var = tk.StringVar()
        entry = tk.Entry(root3, width=20, textvariable=var)
        entry.insert(0, "XXX .xls")
        entry.place(x=230, y=7)
        def save():
            var = entry.get()
            spath = filedialog.askdirectory(title='選擇存檔資料夾')
            rain_all.to_excel(r'{}'.format(os.path.join(spath,str(var))))
            Label=tk.Label(root3,text='沖蝕分析已儲存')
            Label.place(x=100, y=10, width=100, height=30)    
        bsave = tk.Button(root3,text='輸入檔名',command=save)
        bsave.place(x=330, y=5)
        
        LabeL2=tk.Label(root3,text='監測期間共有{}場降雨'.format(len(rain)), font=2)
        LabeL2.place(x=9, y=110, width=200, height=30)
        LabeL3=tk.Label(root3,text='第1筆資料: {}'.format(rain_all.S_time[0]), font=2)
        LabeL3.place(x=10, y=155, width=250, height=35)
        LabeL4=tk.Label(root3,text='第{0}筆資料: {1}'.format(len(rain),rain_all.S_time[len(rain_all.S_time)-1]), font=2)
        LabeL4.place(x=5, y=195, width=270, height=35)

    def Bm():
        root3.destroy()
        root0.deiconify()

    B = tk.Button(root3, text='沖蝕分析', bg='DarkOliveGreen',fg='white', state=tk.NORMAL, command=Erosion)
    B.place(x=10, y=5, width=80, height=30)
    Bm = tk.Button(root3, text='返回主畫面', bg='WhiteSmoke',fg='gray', state=tk.NORMAL, command=Bm)   
    Bm.place(x=10, y=50, width=80, height=30)
    
#====================================================================================主畫面===========================
r2 = tk.Button(root0, text='雨場切割', bg='WhiteSmoke',fg='FireBrick', state=tk.NORMAL, command=createroot2)
r2.place(x=10, y=5, width=120, height=45)
r = tk.Button(root0, text='坡面侵蝕', bg='WhiteSmoke',fg='navy', state=tk.NORMAL, command=createroot)
r.place(x=10, y=60, width=120, height=45)
r3 = tk.Button(root0, text='沖蝕分析', bg='WhiteSmoke',fg='DarkOliveGreen', state=tk.NORMAL, command=createroot3)
r3.place(x=10, y=115, width=120, height=45)

root0.mainloop()
```
