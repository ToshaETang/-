```python
m = 20
s = 2
d = 1
df1 = pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\719\Merge\{0}_ID{1}_d{2}.csv'.format(m,s,d))
df1['Time']=pd.to_datetime(df1['Time'])

for i in range(len(df1)):
    df1['Time'] = df1['Time'].replace(tzinfo=None)
    df1['Time'] = df1['Time'] + timedelta(hours=8)
            
df1 = df1.loc[df1['distance1']!=20]        
df1.set_index('Time', drop=True, inplace=True)
df1=df1.loc[df1["distance1"]>=0]["distance1"]
mean=df1.rolling(30,min_periods=1,center=True).mean()
std=df1.rolling(30,min_periods=1,center=True).std()
i=0
while i < (len(df1)):
    if abs(df1[i]-mean[i])<std[i]:
        i=i+1
    else:
        df1=df1.drop(df1.index[i])
        i=i+1
df2=df1.resample(rule='1D').mean()
mean=df2.rolling(30,min_periods=1,center=True).mean()
std=df2.rolling(30,min_periods=1,center=True).std()
i=0
while i < (len(df2)):
    if abs(df2[i]-mean[i])<std[i]:
        i=i+1
    else:
        df2=df2.drop(df2.index[i])
        i=i+1
df3=df2#.resample(rule='30T').mean().interpolate(method='slinear')
df3.dropna(inplace=True)
df=df3.to_frame()
df.to_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\719\Merge\{0}_ID{1}_d{2}.csv'.format(m,s,d))
print("儲存done")
```
