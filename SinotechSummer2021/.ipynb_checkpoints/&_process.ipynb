```python
import pandas as pd
import numpy as np
import os
import os.path
import datetime,time
import matplotlib.pyplot as plt
import statsmodels
import plotly.express as px
from pandas.plotting import register_matplotlib_converters  
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.seasonal import STL
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
version=str(3.15)
print("process newest version is " +version)

def read(file,save):
    df1 = pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\702\8_3_統計.csv')

    df1=df1.set_index('time')
    df11=df1.drop(['Unnamed: 0','ID','Index','entry_id','field1','field3','field4'],axis=1).rename(columns={'field2':'distance'})
    df12=df1.drop(['Unnamed: 0','ID','Index','entry_id','field1','field2','field4'],axis=1).rename(columns={'field3':'distance'})
        
    df11.to_csv(save + "\{}-1.csv".format(df1.ID[1]))
    df12.to_csv(save + "\{}-2.csv".format(df1.ID[1]))
    print("資料切割完成")
    return df1.ID[1]    

#匯入df1 --> 把df1的time設為索引 --> 刪除time,distance以外的column --> 存成新csv



#df1要匯入切完的資料
df1 = pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\702\8_3-1.csv')
df1 = df1.loc[df1['distance']!=20]

def clean(df1):
    df1['time']=pd.to_datetime(df1['time'])
    df1.set_index('time', drop=True, inplace=True)
    
    df1=df1.loc[df1["distance"]>=30]["distance"]
    mean=df1.rolling(30,min_periods=1,center=True).mean()
    std=df1.rolling(30,min_periods=1,center=True).std()
    i=0
    while i < (len(df1)):
        if abs(df1[i]-mean[i])<std[i]:
            i=i+1
        else:
            df1=df1.drop(df1.index[i])
            i=i+1
    df2=df1.resample(rule='30T').mean()
    mean=df2.rolling(30,min_periods=1,center=True).mean()
    std=df2.rolling(30,min_periods=1,center=True).std()
    i=0
    while i < (len(df2)):
        if abs(df2[i]-mean[i])<std[i]:
            i=i+1
        else:
            df2=df2.drop(df2.index[i])
            i=i+1
    df3=df2.resample(rule='30T').mean().interpolate(method='slinear')
    df3.dropna(inplace=True)
    df=df3.to_frame()
    print("資料清洗完成")
    return df




def analysis(df9):
    res= STL(df9,period=103).fit()
    #print(res.seasonal)
    #res.plot()
    #plt.show()
    #把陣列轉成dataframe
    
    #re3=df9.to_frame()
    #把分解出來的趨勢轉回去
    df9['trend']=res.trend
    df9['season']=res.seasonal
    df9['residual']=res.resid
    #補滿前後資料
    df9=df9.fillna(method='ffill')
    df9=df9.fillna(method='bfill')
    #print(name+"資料STL處理完成")
    df9.to_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\702\df9.csv')
    return df9

#period=


df = pd.read_csv(r'C:\Users\Tosha.E.T\Desktop\Sinotech\702\df9.csv')
print(df)
def Bar(df):
    df.index = pd.to_datetime(pd.Series(df['time']))
    fig = px.bar(df, y='trend', x='time')                
    fig.show()
    print("Done")

    
    
clean(df1)
analysis(df1)    
Bar(df)
```
